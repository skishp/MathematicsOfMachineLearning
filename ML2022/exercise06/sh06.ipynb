````python
# Importing all the required libraries

import numpy as np
import seaborn as sns
import sklearn
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
````

````python
# Reading the Iris dataset
# The iris data set contains 50 samples from three species of Iris (setosa, virginica and versicolor) 
# and their sepal length, sepal width, petal length and petal width

iris=sklearn.datasets.load_iris()

# DESCR gives us a discription of the data set

print(iris.DESCR)
````

````python
#Splitting the dataset into the training and testing set

X=iris.data
y=iris.target

x_train, x_test, y_train, y_test= train_test_split (X,y)
````

# Exercise 3

Visualize the petal length and sepal length of the training set in a 2D plot: Plot the petal length on the x axis, the sepal length on the y axis and use colors for the classes.

````python
# for visualization we use the library matplotlib.pyplot (which we already imported at the beginning when we wrote import matplotlib.pyplot as plt)

# use scatter plot

# add labels to the x and y axis
y_axis = x_train[:,0]
x_axis = x_train[:,2]

print(y_axis)

plt.scatter(x_axis, y_axis,c=y_train)
plt.xlabel("petal length")
plt.ylabel("sepal length")
````

# After visualizing the data, now lets move on to train a Decission Tree!

````python
#Defining our model

model = tree.DecisionTreeClassifier()
````

````python
# We already split the data set in a training set and test set at the beginning
# Now lets fit our model for the training set

model.fit(x_train,y_train)
````

````python
#Plotting our model in the form of a tree

plt.figure(figsize=(15,15))
tree.plot_tree(model)
plt.show()
````

````python
y_pred = model.predict(x_train)

accuracy_score(y_train,y_pred)
````

So the training error is 0, which is no surprise since we didn't bound the depth of the decission tree and it was trained until it perfectly matched the data.

````python
#Predicting the value of the test data
y_pred = model.predict(x_test)
y_pred
````

````python
#Computing the accuracy of our model

accuracy_score(y_test,y_pred)
````

We would normally expect overfitting, since the decission tree is really complex compared to the data with four features, but our model performs well on the test data.

# Exercise 4

Assume we lost the data regarding the sepal width and petal width. 

First, create a new data set containing only the sepal length, petal length and the class. Split the data set in training and test set. Train a decission tree on the training set. How is it's accuracy on the training set?

````python
#since it was not specified how the the data looks
#we will be using np random to create data set
x_data_sepal = np.random.rand(1000,1)
x_data_sepal = x_data_sepal * 7


x_data_petal  = np.random.rand(1000,1)
x_data_petal = x_data_petal*7

#concatenate with axis=1 to get right data format 
x_data  = np.concatenate((x_data_sepal,x_data_petal), axis=1)

#create labels for data
y_data  = np.array([np.random.randint(0,3,1000)])

#rotate data to match data formats
y_data = np.rot90(y_data)

#splitting data set into train/test set
x_train, x_test, y_train, y_test = train_test_split(x_data,y_data)

#define model 
model = tree.DecisionTreeClassifier()

#train model 
model.fit(x_train,y_train)

#predict on train set
y_predict = model.predict(x_train)

#Training error is still 0, so accuracy score is 1
print(accuracy_score(y_train, y_predict))
````
